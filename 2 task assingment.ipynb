# Introduction

Python is a high-level, easy-to-read programming language designed to make coding simple and efficient. It was created in 1991 and is widely used in **economics, finance, and research**.


##Key Concepts

- Python uses simple syntax  
- It is interpreted  
- Supports object-oriented programming  
- Useful libraries: NumPy, Pandas, Matplotlib  


# Simple Syntax Examples


```python
x = 10
numbers = [1, 2, 3]

if x > 5:
    print("Greater")

for i in numbers:
    print(i)
```
